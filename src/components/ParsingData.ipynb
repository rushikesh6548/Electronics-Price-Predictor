## This code describes two sets of HTML pages:
### - The first set contains almost 24 laptops per page, with prices ranging from Rs 4,79,990 to almost Rs 57,000.
### - The second set also contains laptops ranging from Rs 20,000 to 60,000 per page.
#### This way we get all the laptops that are available 

### First we parse data from first set  

```python
from bs4 import BeautifulSoup as bs 
import requests

def get_laptop_names(html_content):
    soup = bs(html_content, 'html.parser')
    laptop_names = []
    for laptop in soup.find_all('div', {'class': '_4rR01T'}):
        laptop_names.append(laptop.text)
    return laptop_names

# Extract the laptop names
laptop_names_set1 = []
for i in range(1, 41):
    with open(f'{i}_flipkart_laptop.html', 'r',encoding='utf-8') as f:
        html_content = f.read()
    laptop_names_set1 += get_laptop_names(html_content)

print(laptop_names_set1)
```

```python
# Extracting Price from all the html pages : 
def get_laptop_prices(html_content):
    soup = bs(html_content, 'html.parser')
    laptop_prices = []
    for laptop in soup.find_all('div', {'class': '_30jeq3 _1_WHN1'}):
        laptop_prices.append(laptop.text)
    return laptop_prices

# Extract the laptop prices
laptop_prices_set1 = []
for i in range(1, 41):
    with open(f'{i}_flipkart_laptop.html', 'r',encoding='utf-8') as f:
        html_content = f.read()
    laptop_prices_set1 += get_laptop_prices(html_content)

print(laptop_prices_set1)

```
